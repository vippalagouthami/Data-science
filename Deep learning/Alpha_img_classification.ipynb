# Scaled image

```python
X_train_scale=X_train/255.0
```

```python
X_test_scale=X_test/255.0
```

```python
from tensorflow.keras.models import Model
```

```python
from tensorflow.keras.layers import Input 
```

```python
feed=Input(shape=(28,28,3))
```

```python
flat=Flatten()(feed)
```

```python
d1=Dense(units=512,activation="relu")(flat)
```

```python
d2=Dense(units=256,activation="relu")(d1)
```

```python
d3=Dense(units=128,activation="relu")(d2)
```

```python
d4=Dense(units=64,activation="relu")(d3)
```

```python
d5=Dense(units=128,activation="relu")(d4)
```

```python
d6=Dense(units=256,activation="relu")(d5)
```

```python
d7=Dense(units=512,activation="relu")(d6)
```

```python
out=Dense(units=26,activation="softmax")(d7)
```

```python
fmodel=Model(inputs=feed,outputs=out)
```

```python
fmodel.summary()
```

```python
fmodel.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
fmodel.fit(X_train_scale, y_train, epochs = 20)
```

```python
fmodel.evaluate(X_test_scale, y_test)
```
