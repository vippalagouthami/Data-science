```python
X=dfdia.drop("Outcome",axis=1)
```

```python
y=dfdia["Outcome"]
```

```python
dfdia["Outcome"].value_counts()
```

```python
 from imblearn.over_sampling import RandomOverSampler
```

```python
ros=RandomOverSampler()
```

```python
X_ros,y_ros=ros.fit_resample(X,y)
```

```python
X.shape,X_ros.shape,y.shape,y_ros.shape
```

```python
y_ros.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_ros,y_ros,test_size=0.20,random_state=2)
```

```python
X.shape,X_train.shape,X_test.shape
```

```python
y.shape,y_train.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBRFClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier
dtree=DecisionTreeClassifier()
svm=SVC()
knn=KNeighborsClassifier()
xgb=XGBRFClassifier()
rfc=RandomForestClassifier()
ada=AdaBoostClassifier()
grad=GradientBoostingClassifier()
mlist=[dtree,svm,knn,xgb,rfc,ada,grad]
```

```python
mlist
```

```python
for i in mlist:
    print(i,end="==>")
    i.fit(X_train,y_train)
    m=i.score(X_test,y_test)
    print(m)
```

```python
rlist=[]
for i in range (0,51):
    print("random_state=",i)
    X_train,X_test,y_train,y_test=train_test_split(X_ros,y_ros,test_size=0.20,random_state=i)
    for m in mlist:
        print(m,end="==>")
        m.fit(X_train,y_train)
        s=m.score(X_test,y_test)
        print(s)
        rlist.append(s)
    print(rlist) 
    print("________________________________________________________________________")
    print("Max =",max(rlist)) 
    rlist.index(max(rlist))
    print("Max_index=",rlist.index(max(rlist)))
    print("__________________________________________________________________________")
```
